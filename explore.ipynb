Let's determine the spaces of cuisines and ingredients

```python
from collections import Counter
ingredient_counts = Counter()
cuisine_counts = Counter()
for recipe in recipes_json:
    cuisine_counts[recipe['cuisine']] += 1
    for ingredient in recipe['ingredients']: ingredient_counts[ingredient] += 1
```

```python
print("There are {:d} cuisines and {:d} ingredients.".format(len(cuisine_counts),len(ingredient_counts)))
```

```python
print("The cuisine counts are:")
for cuisine in sorted(cuisine_counts.keys()): 
    print("{:s}: {:d}".format(cuisine, cuisine_counts[cuisine]))
```

```python
print("The ingredient counts are:")
for ingredient in sorted(ingredient_counts.keys()): 
    print("{:s}: {:d}".format(ingredient, ingredient_counts[ingredient]))
```

The ingredient names are fairly messy. Brand names and product names abound (often accommpanied by special markers such as copyright symbols). Qualifiers with debateable amounts information also appear (diced, sliced, light/lite, frozen) as well as items with multiple names (catsup/ketchup).

Let's get this data into data frames for some more detailed analyis. The target info with the recipe id and cuisine we'll store in one frame and the feature data in another. Start with a melted representation of the feature data in a two column table of id-ingredient pairs, so each id appears several times.

```python
recipe_cuisine = pd.DataFrame([{'id':recipe['id'], 'cuisine':recipe['cuisine']} for recipe in recipes_json], 
                              columns=["id","cuisine"])
recipe_ingredients = pd.DataFrame([{'id':recipe['id'], 'ingredient':ingredient} for recipe in recipes_json for ingredient in recipe['ingredients']],
                                  columns=['id','ingredient'])
```

```python
print("Minimum number of ingredients is {:d}.".format(recipe_ingredients.groupby('id').size().min()))
print("Maximum number of ingredients is {:d}.".format(recipe_ingredients.groupby('id').size().max()))
```

Turns out the size of the ingredient lists vary a lot. There are a surprising number of recipes with a single ingredient and the longest list tops out at 65 ingredients.
